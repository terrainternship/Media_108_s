## Функции


```python
# Функция для ввода ключа OpenAI API
def get_openai_api_key():
  openai_key = getpass.getpass('OpenAI API Key:')
  os.environ['OPENAI_API_KEY'] = openai_key
  openai.api_key = openai_key


# Функция для подсчёта токенов ChatOpenAI
def get_num_tokens(prompt, question, temperature=0):

  messages = [SystemMessage(content=prompt),
              HumanMessage(content= f'Вопрос пользователя: {question}')]
  chat     = ChatOpenAI(temperature=temperature)

  return chat.get_num_tokens_from_messages(messages)
```
