```python
# Заполним добавленную колонку.
name_to_dir = {'Павелецкая (pavcity.turbo.site).csv':'Записи звонков_павелецкая сити',
               'Headliner.csv':'Записи звонков_хедлайнер',
               'Primavera .csv':'Записи звонков_primavera',
               'Павелецкая_ЖК_«Павелецкая_сити».csv':'Записи звонков_павелецкая сити'}   # Словарь соответствия между файлами инф. о звоках и деррикториями с записями.
duofile = 0
for i in range(len(df)):
    namedir = name_to_dir[df[i]['Name'][0]]
    for files in os.listdir(path_dir+namedir):
        session_in_file = re.findall(r'session_(\d*)_talk',files)
        for k in range(df[i].shape[0]):
            session = str(df[i]['Идентификатор сессии звонка'][k])
            if session == str(session_in_file[0]):
                if df[i].loc[k,['Имя файла']].isna:
                    df[i].loc[k,['Имя файла']] = namedir+'/'+files
                else:
                    nowfile = df[i].loc[k,['Имя файла']]
                    df[i].loc[k,['Имя файла']] = nowfile+'\n'+namedir+'/'+files
                    duofile +=1
print(f'Заполнение завершено. Дублей идентификаторов - {duofile}')
```

Speech_to_text:  попробуем распознать при помощи библиотеки whisper без очистки от шумов

```python
# Результаты будем сохранять после нескольких распознаний, чтобы не потерять результаты.
!pip install openai-whisper
import whisper

```

```python
def speech_recognition(model='base',namefile='Non'):
    if namefile == 'Non':
        raise KeyError('Не передан файл для обработки!')

    spch_mod = whisper.load_model(model)
    return spch_mod.transcribe(namefile,)['text']
```

```python
# Пройдемся по нашим "новым" датафреймам и траскрибируем аудио файлы
# Датафреймы в которых будем сохранять распознанный текст и его признаки
df_head_txt = pd.DataFrame(columns=['Идентификатор сессии звонка', 'Теги', 'Текст'])
df_prima_txt = pd.DataFrame(columns=['Идентификатор сессии звонка', 'Теги', 'Текст'])
path_dir = '/content/drive/MyDrive/Media108/DS/'
for i in range(df_head.shape[0]):
    filename = path_dir+df_head.iloc[i,0]
    id_session = df_head.iloc[i,df_head.columns.get_loc('Идентификатор сессии звонка')]
    tegs = df_head.iloc[i,df_head.columns.get_loc('Теги')]
    rezalt = speech_recognition(model='base',namefile=filename)
    rowadd = {'Идентификатор сессии звонка':id_session, 'Теги':tegs, 'Текст':rezalt}
    df_head_txt = df_head_txt.append(rowadd, ignore_index=True)
    with open('/content/drive/MyDrive/Media108/Head_txt.pkl', 'wb') as f:
        pkl.dump(df_head_txt, f)
```
